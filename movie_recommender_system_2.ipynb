```python
# This ast is used for formatting the list without string
import ast
```

```python
def convert (obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L  
```

```python
# we will keep top three actors from the cast column
def convert(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter += 1
    else:
      break

  return L        
        
```

Find the only director from the dictionary

```python
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break

  return L
```

**Text vecotrization**
It's can be done by three way
1. Bag of words
2. Tf idf
3. word to vect

we will use 1. bag of words

For implementing the vectorization, at first we need to remove the stop word from the sentence.

Stemmer is used for convert the same kinds of word into one. 
like ['love', 'loves','loving','loved'] = ['love']

```python
!pip install nltk
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)  
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

We will calculate cosine distance but not uclidian distance
1. cosine distance is angular distance 
2. uclidian distance top part to top part

N.B: Since we work on high dimensional data, So uclidian will not work and fail the program.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
# per movie compare distance with each other 
# So, 4806,4806 ---- shape
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
# first movie compare the distance with others
# distance first movie with first movie will 1 
# also for everyone with everyone will 1
similarity[0]
```

```python
list(enumerate(similarity[0]))
# it creates a touple like zero movier sathe zero movier distance koto,
# one index movier sathe distance koto etc...
```

```python
sorted(list(enumerate(similarity[0])),reverse=True)
# this sorting is index based,
# but we need to second one that means distance base
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
# take five movies from the distance based sorting 
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distance = similarity[movie_index]
  movies_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
  
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
    #print(i[0])
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
